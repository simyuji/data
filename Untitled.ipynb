```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
happy=pd.read_csv("Happiness.csv", sep=",")
happy.head()
```

```python
happy.groupby('Region').Beer_PerCapita.mean()
```

```python
happy.groupby('Region').Wine_PerCapita.describe()
```

```python
happy.groupby('Region').mean()
```

```python
import seaborn as sns
plt.figure(figsize = (15,6))
sns.barplot(x ='HappinessScore',y= 'Region',data = happy, palette = 'viridis')
plt.title('Happiness Score by Region', fontsize = 20)
plt.xticks(rotation = 90)
plt.show()
```
